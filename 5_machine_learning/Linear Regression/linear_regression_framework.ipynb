```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split

study_hours = np.random.uniform(2, 10, 50) 
# Reshape study_hours to a 2D array, as scikit-learn expects features to be 2D
X = study_hours.reshape(-1, 1)
exam_scores = 40 + 6 * study_hours + np.random.normal(0, 5, 50)
y = exam_scores # Target variable

print("Sample of Study Hours (X):\n", X[:5])
print("\nSample of Exam Scores (Y):\n", y[:5])

# Visualize the data
plt.figure(figsize=(8, 6))
plt.scatter(X, y, color='blue', alpha=0.7, label='Actual Data Points')
plt.xlabel("Study Hours (X)")
plt.ylabel("Exam Score (Y)")
plt.title("Study Hours vs. Exam Score (Scikit-learn)")
plt.grid(True)
plt.legend()
plt.show()
```

```python
model = LinearRegression()
model.fit(X, y)
```

```python


slop_m = model.coef_[0]


b = model.intercept_

print("slop_m : ",slop_m)
print("b : ",b)
```

```python
# prediction
def predict(x_values, b, slop_m):
    return b + slop_m * x_values


y_predicted = model.predict(X)

print("\nSample of Actual Exam Scores:", y[:5])
print("Sample of Predicted Exam Scores (Scikit-learn):", np.round(y_predicted[:5], 2))
```

```python
mse_sklearn = mean_squared_error(y, y_predicted)
print(f"Mean Squared Error (MSE): {mse_sklearn:.2f}")
```

```python
new_study_hour = np.array([6.5]) # [6.5]
#new_study_hour= new_study_hour.reshape(-1, 1)
print("new_study_hour : ",new_study_hour)
predicted_score_new = model.predict([new_study_hour])
print("predicted_score_new : ",predicted_score_new)
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(study_hours, exam_scores, color='blue', label='Actual Data Points')
plt.plot(study_hours, y_predicted, color='red', label='Regression Line')
plt.scatter(new_study_hour, predicted_score_new, color='green', marker='X', s=100, label=f'Predicted ({new_study_hour}hrs)')
plt.xlabel("Study Hours (X)")
plt.ylabel("Exam Score (Y)")
plt.title("Simple Linear Regression: Best-Fit Line")
plt.grid(True)
plt.legend()
plt.show()
```
