## Linear Regression 


**Equation of a Line:** $y = mx + b$

**The Goal of Training:** Find the values of $b$ and $m$ that **minimize the MSE**.



**Slop formula (m) :** (y2 - y1) /(x2 - x1)

```python
import numpy as np
import matplotlib.pyplot as plt


study_hours = np.random.uniform(2, 10, 50) 
exam_scores = 40 + 6 * study_hours + np.random.normal(0, 5, 50)


plt.figure(figsize=(6, 4))
plt.scatter(study_hours, exam_scores, color='blue', label='Actual Data Points')
plt.xlabel("Study Hours (X)")
plt.ylabel("Exam Score (Y)")
plt.title("Study Hours vs. Exam Score")
plt.grid(True)
plt.legend()
plt.show()
```

```python
mean_x = np.mean(study_hours)
mean_y = np.mean(exam_scores)
n = len(study_hours)

print("Mean X:", mean_x)
print("Mean Y:", mean_y)
print("Number of data points:", n)
```

```python
# Calculate numerator for slop(m)
# Σ(xi - x̄)(yi - ȳ)
numerator = np.sum((study_hours - mean_x) * (exam_scores - mean_y))

# Calculate denominator for slop(m)
# Σ(xi - x̄)^2
denominator = np.sum((study_hours - mean_x)**2)


slop_m = numerator / denominator #(y2-y1)/(x2-x1)


b = mean_y - slop_m * mean_x

print("slop_m : ",slop_m)
print("b : ",b)
```

```python
# prediction
def predict(x_values, b, slop_m):
    return b + slop_m * x_values


y_predicted = predict(study_hours, b, slop_m)
print("y_predicted : ",y_predicted)
```

```python
print("y_true : ",exam_scores)
```

```python
mse = np.mean((exam_scores - y_predicted)**2)
print(f"Mean Squared Error (MSE): {mse:.2f}")
```

```python
new_study_hour = 4
predicted_score_new = predict(new_study_hour, b, slop_m)
print("predicted_score_new : ",predicted_score_new)
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(study_hours, exam_scores, color='blue', label='Actual Data Points')
plt.plot(study_hours, y_predicted, color='red', label='Regression Line')
plt.scatter(new_study_hour, predicted_score_new, color='green', marker='X', s=100, label=f'Predicted ({new_study_hour}hrs)')
plt.xlabel("Study Hours (X)")
plt.ylabel("Exam Score (Y)")
plt.title("Simple Linear Regression: Best-Fit Line")
plt.grid(True)
plt.legend()
plt.show()
```
