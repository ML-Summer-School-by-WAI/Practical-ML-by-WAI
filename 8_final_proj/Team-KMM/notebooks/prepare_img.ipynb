```python
import os
import cv2
import random
from dotenv import load_dotenv
import matplotlib.pyplot as plt
```

```python
# Load environment variables
load_dotenv()
ML_Summer_School_ID = os.getenv('ML_Summer_School_ID')
print("Your Student ID is:", ML_Summer_School_ID)
```

```python
# Base paths
project_root = os.path.abspath(os.path.join(os.getcwd(), ".."))  # Go one level up
base_dir = os.path.join(project_root, "dataset")
cats_folder = os.path.join(base_dir, "cat_and_dog_dataset", "cats")
dogs_folder = os.path.join(base_dir, "cat_and_dog_dataset", "dogs")
annotations_folder = os.path.join(base_dir, "annotations")
```

```python
# Function to get image files only
def get_image_files(folder):
    return [f for f in os.listdir(folder) if f.lower().endswith(('.png', '.jpg', '.jpeg'))]

cats_images = get_image_files(cats_folder)
dogs_images = get_image_files(dogs_folder)

print(f"Number of cat images: {len(cats_images)}")
print(f"Number of dog images: {len(dogs_images)}")
```

```python
# Display a random cat image
cat_img = random.choice(cats_images)
cat_path = os.path.join(cats_folder, cat_img)
cat_img_data = cv2.imread(cat_path)
cat_img_resized = cv2.resize(cat_img_data, (300, 300))

plt.imshow(cv2.cvtColor(cat_img_resized, cv2.COLOR_BGR2RGB))
plt.title(f"Cat: {cat_img}")
plt.axis('off')
plt.show()
```

```python
# Save cat image
cat_save_path = os.path.join(annotations_folder, f"{ML_Summer_School_ID}_cat.png")
cv2.imwrite(cat_save_path, cat_img_resized)
print(f"Saved cat image at: {cat_save_path}")
```

```python
# Display a random dog image
dog_img = random.choice(dogs_images)
dog_path = os.path.join(dogs_folder, dog_img)
dog_img_data = cv2.imread(dog_path)
dog_img_resized = cv2.resize(dog_img_data, (300, 300))

plt.imshow(cv2.cvtColor(dog_img_resized, cv2.COLOR_BGR2RGB))
plt.title(f"Dog: {dog_img}")
plt.axis('off')
plt.show()
```

```python
# Save dog image
dog_save_path = os.path.join(annotations_folder, f"{ML_Summer_School_ID}_dog.png")
cv2.imwrite(dog_save_path, dog_img_resized)
print(f"Saved dog image at: {dog_save_path}")
```
