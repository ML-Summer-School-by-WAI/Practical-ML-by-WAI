```python
import cv2
import os
from dotenv import load_dotenv

load_dotenv()

ML_Summer_School_ID = os.getenv('ML_Summer_School_ID')
print("Your Sudent ID is: " + ML_Summer_School_ID)
```

```python
import matplotlib.pyplot as plt

# Define the paths for your two image folders
cats = './cats_and_dogs_filtered/train/cats'
dogs = './cats_and_dogs_filtered/train/dogs'
```

```python
# A list to store the full paths of the images to be processed
images_to_process = []

# Process the 'cat' folder from cat.0.jpg to cat.90.jpg
for i in range(91):
    img_name = f'cat.{i}.jpg'
    img_path = os.path.join(cats, img_name)
    if os.path.exists(img_path):
        images_to_process.append(img_path)
    else:
        print(f"Warning: Image not found at {img_path}. Skipping.")
```

```python
# Process the 'dog' folder from dog.0.jpg to dog.90.jpg
for i in range(91):
    img_name = f'dog.{i}.jpg'
    img_path = os.path.join(dogs, img_name)
    if os.path.exists(img_path):
        images_to_process.append(img_path)
    else:
        print(f"Warning: Image not found at {img_path}. Skipping.")
```

```python
print(f"Found {len(images_to_process)} images to process.")
```

```python
# A variable to hold the last resized image
last_resized_img = None

# Process each image in the list
for img_path in images_to_process:
    print(f"Processing image: {img_path}")
    
    # Read the image
    read_img = cv2.imread(img_path)
    
    # Check if the image was read successfully
    if read_img is None:
        print(f"Image not found at {img_path}")
        continue
    
    # Resize the image to 300x300
    last_resized_img = cv2.resize(read_img, (300, 300))
    
    # The original code only showed the plot for one image.
    # To see all of them, you would need to plot inside the loop,
    # but that can be resource-intensive for many images.

# Display the last resized image
if last_resized_img is not None:
    plt.imshow(last_resized_img)
    plt.title("Last Resized Image")
    plt.show()
else:
    print("No images were processed to display.")
```
