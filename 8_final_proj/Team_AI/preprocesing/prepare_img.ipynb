```python
import random
import matplotlib.pyplot as plt

test_img_path = "./JPEGImages(50-74)"
os.listdir(test_img_path)
```

```python
selected_img = random.choice(os.listdir(test_img_path))
print("selected_img : " + selected_img)



selected_img_full_path = test_img_path+"/"+selected_img

print("selected_img_full_path : ",selected_img_full_path)

read_img = cv2.imread(selected_img_full_path)
if read_img is None:
    print("Image not found")


resize_img = cv2.resize(read_img,(300,300))
plt.imshow(resize_img)
```

```python
selected_img = "dog.74.jpg"
print("selected_img : " + selected_img)

selected_img_full_path = os.path.join(test_img_path, selected_img)
print("selected_img_full_path : ", selected_img_full_path)

read_img = cv2.imread(selected_img_full_path)
if read_img is None:
    print("Image not found")

resize_img = cv2.resize(read_img, (300, 300))
plt.imshow(cv2.cvtColor(resize_img, cv2.COLOR_BGR2RGB))
```

```python
filename_no_ext = os.path.splitext(selected_img)[0]  # remove extension
write_name = f"./data_annotated/{filename_no_ext}.png"
cv2.imwrite(write_name, resize_img)

print("Saved as:", write_name)
```

```python
folder = "./data_annotated"

for fname in os.listdir(folder):
    if fname.lower().endswith(".png"):
        png_path = os.path.join(folder, fname)
        jpg_path = os.path.splitext(png_path)[0] + ".jpg"

        img = cv2.imread(png_path)
        cv2.imwrite(jpg_path, img, [int(cv2.IMWRITE_JPEG_QUALITY), 95])
        print("Converted:", jpg_path)
```
