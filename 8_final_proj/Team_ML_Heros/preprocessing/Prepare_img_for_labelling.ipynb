```python
import cv2
import os
import random
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Path to the dataset
image_path = './cats_and_dogs_filtered/train'
dog_image_path = os.path.join(image_path, 'dogs')
cat_image_path = os.path.join(image_path, 'cats')
```

```python
# Counting number of images in each category
number_of_files_dogs = len(os.listdir(dog_image_path))
print(f'Number of dog images: {number_of_files_dogs}')
```

```python
number_of_files_cats = len(os.listdir(cat_image_path))
print(f'Number of cat images: {number_of_files_cats}')
```

```python
# Reading random dog image
random_dog_image = random.choice(os.listdir(dog_image_path))
dog_img = cv2.imread(os.path.join(dog_image_path, random_dog_image))
if dog_img is None:
    print("Image not found or unable to load.")

resized_dog_img = cv2.resize(dog_img, (224, 224))
plt.imshow(cv2.cvtColor(resized_dog_img, cv2.COLOR_BGR2RGB))
```

```python
# Reading random cat image
random_cat_image = random.choice(os.listdir(cat_image_path)) 
cat_img = cv2.imread(os.path.join(cat_image_path, random_cat_image))
if cat_img is None:
    print("Image not found or unable to load.") 
resized_cat_img = cv2.resize(cat_img, (224, 224))
plt.imshow(cv2.cvtColor(resized_cat_img, cv2.COLOR_BGR2RGB))
```
